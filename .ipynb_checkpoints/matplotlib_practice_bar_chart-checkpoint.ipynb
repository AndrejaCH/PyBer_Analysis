```python
#Magic command (%): setting back-end processor for charts inline: displays chart within the cell for a Jupyter Notebook.
%matplotlib inline
```

```python
#Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Create the plot - data on the opposite axes - not a good option for this grap, but we can do it!
plt.barh(y_axis, x_axis)
```

```python
#INVERT THE ORDER ON AXIS
#gca() method = "get current axes." To invert the y-axis in ascending order.
#chain the gca() to the invert_yaxis(): gca().invert_yaxis()

# Create the plot.
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
#SKILL-DRILL:
# Create the plot.
plt.barh(x_axis, y_axis, color="magenta", label='Boston')

#invert the y_axis
plt.gca().invert_yaxis()

# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Date")

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
#USING OBJECT-ORIENTED METHOD:
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
#SKILL-DRILL:
# Create the plot.
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')

#invert the y_axis
ax.invert_yaxis()

# Create labels for the x and y axes.
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

# Create a title.
ax.set_title("PyBer Fare by Month")

# Add the legend.
ax.legend()
```
