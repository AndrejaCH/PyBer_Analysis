# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Read & Load files 
city_data_df = pd.read_csv("Resources/city_data.csv")
ride_data_df = pd.read_csv("Resources/ride_data.csv")
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_per_city_type = pyber_data_df.groupby(['type']).count()['ride_id']
total_rides_per_city_type
```

```python
# 2. Get the total drivers for each city type
total_drivers_per_city_type = city_data_df.groupby(['type']).sum()['driver_count']
total_drivers_per_city_type
```

```python
#  3. Get the total amount of fares for each city type
total_fares_per_city_type = pyber_data_df.groupby(['type']).sum()['fare']
total_fares_per_city_type
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride = pyber_data_df.groupby(['type']).sum()['fare'] / total_rides_per_city_type
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = pyber_data_df.groupby(['type']).sum()['fare'] / total_drivers_per_city_type
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({"Total Rides": total_rides_per_city_type,
                                 "Total Drivers" :total_drivers_per_city_type,
                                 "Total Fares" : total_fares_per_city_type,
                                 "Average Fare per Ride" : average_fare_per_ride,
                                 "Average Fare per Driver": average_fare_per_driver})

#pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,.0f}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,.0f}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 9. Using groupby() on the pyber_data_df to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
pyber_total_fares_summary_df = pyber_data_df.groupby(['date', 'type']).sum()[['fare']]
pyber_total_fares_summary_df.head()
```

```python
# 10. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
pyber_total_fares_summary_df = pyber_total_fares_summary_df.reset_index()
pyber_total_fares_summary_df.head()
```

```python
# 11. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_total_fares_summary_df = pyber_total_fares_summary_df.pivot(index="date", columns="type", values="fare")
pyber_total_fares_summary_df.head(10)
```

```python
# 12. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2018-01-01':'2018-04-29'.
pyber_total_fares_summary_df = pyber_total_fares_summary_df.loc['2019-01-01':'2019-04-29', :]
pyber_total_fares_summary_df
```

```python
# 13. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

pyber_total_fares_summary_df.index = pd.to_datetime(pyber_total_fares_summary_df.index)
pyber_total_fares_summary_df
```

```python
# 14. Check that the datatype for the index is datetime using df.info()
pyber_total_fares_summary_df.info()
```

```python
# 15. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pyber_total_fares_summary_week_df = pyber_total_fares_summary_df.resample("W").sum()
pyber_total_fares_summary_week_df
```

```python
# 16. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

#Assigning xticks and defining "date" column as index
ax = pyber_total_fares_summary_week_df.plot(xticks=pyber_total_fares_summary_week_df.index, figsize=(15, 6))

#Name of the axis and title
ax.set_xlabel(None)
ax.set_ylabel("Fare($USD)")
ax.set_title("Total Fare by City Type")

# Save the figure.
plt.savefig("Analysis/PyBer_fare_summary.png")

plt.show()
```
